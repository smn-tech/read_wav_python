```python
import numpy as np
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
```

**Load audio data**

```python
file_name='audio_data.wav'
Fs,data=read(file_name)

print('[=] Sampling rate {} Hz'.format(Fs))

if len(data.shape)==2:
    print('[=] Stereo channel')
elif len(data.shape)==1:
    print('[=] Mono channel')
print(type(data))
```

**Read channel data**

```python
data_lch=data[:,0]
data_rch=data[:,1]

N=data_lch.size
Ts=1/Fs
t=np.arange(N)*Ts
```

**plot data**


```python
plt.figure(figsize=(12,4))

plt.subplot(2,1,1)
plt.plot(t,data_lch,color=[0,0,0],linewidth=0.5)
plt.xticks(ticks=[])
plt.ylabel('Left chh')
plt.grid()

plt.subplot(2,1,2)
plt.plot(t,data_rch,color=[0,0,0],linewidth=0.5)
plt.xlabel('Time (s)')
plt.ylabel('Right chh')
plt.grid()

plt.show()
```
